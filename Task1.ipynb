<h1>Data Collection</h1>
we will define a function for retrieving data from the JCDecaux Bikes API,We are using contract(city) as key to search and find the datas

```python
def fetch_bike_stations_by_contract(api_key, contract):
    endpoint = 'https://api.jcdecaux.com/vls/v3/stations'
    params = {
        'apiKey': api_key,
        'contract': contract,
    }
    query_string = urllib.parse.urlencode(params)
    url = f"{endpoint}?{query_string}"
    with urllib.request.urlopen(url) as response:
        if response.status == 200:
            return json.loads(response.read().decode())
        else:
            raise Exception(f"HTTP error: {response.status}")
```

we will get all of information from JCDecaux Bikes API by fetch_bike_stations_by_contract function.So I define a new function to store this in local file as json.

```python
def fetch_storeJson(city_name):
    stations = fetch_bike_stations_by_contract(api_key,city_name)
    fname ="%s" % city_name
    out_path=dir_raw / fname
    print("Writing data to %s" % out_path)
    fout = open(out_path, "w")
    json.dump(stations, fout, indent=4, sort_keys=True)
    fout.close()
```

Get data for each region by looping through by fetch_storeJson function

```python
for i in city_names:
    fetch_storeJson(i)
```
